## Introduction to Chemical Machine Learning 

In this Notebook we are going to load a dataset of molecules for which we have a measured property. We will collect some molecular descriptors and visualize how the target property varies across the feature space.

The overall workflow involves the following steps:
- Conversion of SMILES to 2D and 3D structures (rdkit)
- Obtaining 2D and 3D molecular descriptors (rdkit)
- Populating a tabular dataframe with the y-values and the descriptors (pandas)
- Dimensionality reduction using PCA and t-SNE to visualize the chemical space (scikit-learn)
- Clustering of the data with k-means (scikit-learn)

#### 1. Import various python modules

```python
# import python modules
import numpy as np
import pandas as pd

# scikit learn!
import sklearn
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# nice plotting
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns; sns.set()

# for 3d plotting
from mpl_toolkits.mplot3d import Axes3D

# for progress bars
from tqdm import tqdm
```

#### 2. Import rdkit and functions
We will use rdkit to convert SMILES to chemical structures, to generate 3D conformations, and to obtain chemical descriptors

```python
# import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import rdkit
# record which version of the rdkit we are using
print(rdkit.__version__)
```

#### 3. Read in a CSV file containing a dataset of molecules and a measured property that we want to model. 
In this case, the data comes from the following paper: *J. Chem. Inf. Model.* **2014**, *54*, 1604−1616 (DOI: 10.1021/ci5001778)

"386 chemists voted on the molecular complexity (on a scale of 1−5) of molecules taken from various sources: public, licensed, and in-house. The *meanComplexity* of a molecule is the average over all votes for that molecule"

```python
# Define path and csv files 
DATA_DIR = './'
INPUTS = 'ci5001778_si_001.csv'

# Load the raw data as a pandas dataframe
data = pd.read_csv(DATA_DIR + INPUTS)
data.head()

# remove duplicate entries
data = data.drop_duplicates(subset=['SMILES'])
```

```python
print(data.shape)
```

The dataset has 1775 entries with 6 columns

#### 4. Convert the SMILES into molecular structures

One of RDKit's most useful classes is the 'Mol' (molecule) class. This stores the structure of a molecule and facilitates the calculation of various atomic and molecular properties. A convenient way to generate a Mol object is to use the RDKit function MolFromSmiles(), which takes as an input a SMILES string and returns the Mol object associated with that molecule.

```python
mols = data['SMILES']
mol_objects = [Chem.MolFromSmiles(mol) for mol in mols]
```

```python
#Drop rows where rdkit failed to generate mol objects

data["mol objects"] = mol_objects
print(data.shape)
data_formatted = data[data['mol objects'].notna()]
print(data_formatted.shape)

mol_objects = data_formatted["mol objects"]
data = data.drop(['mol objects'], axis=1)
```

We lost 12 entries from the original dataset due to invalid SMILES

```python
# show the first 6 molecules in our dataset
Draw.MolsToGridImage(mol_objects[0:6],subImgSize=(300, 250))
```

#### 5. Molecular Descriptors
So far we have a set of molecular structures which each has an associated y value / property that we want to model. 

The traditional paradigm of cheminformatics is to now obtain a set of molecular descriptors that enable us to represent each molecule in a vector space (where the number of descriptors = the dimensionality of that space). 

###### Let's take a look at some of the molecular descriptors RDKit has available. These descriptors can be grouped according to the structural information that they encode:

1D: count descriptors, e.g. number of rings
2D: topological descriptors are based on the connectivity (i.e., bonding information) and encode the 2D molecular graph
3D: descriptors that encode information of molecular shape, conformation
4D: descriptors that encode information of molecular shape across the conformer ensemble.

```python
from rdkit.Chem import rdMolDescriptors, Descriptors, GraphDescriptors, Descriptors3D
```

##### Count Descriptors:
https://rdkit.org/docs/source/rdkit.Chem.Descriptors.html

```python
descriptor_df = pd.DataFrame()
descriptor_df['SMILES'] = data_formatted['SMILES']
descriptor_df['Mol Weight'] = [Descriptors.MolWt(mol) for mol in mol_objects]
descriptor_df['num Aromatic Rings'] = [rdMolDescriptors.CalcNumAromaticRings(mol) for mol in mol_objects]
descriptor_df['num Stereo Centers'] = [rdMolDescriptors.CalcNumAtomStereoCenters(mol) for mol in mol_objects]

print(descriptor_df.shape)
descriptor_df.head()
```

##### Topological Descriptors
https://www.rdkit.org/new_docs/source/rdkit.Chem.GraphDescriptors.html

Balaban: *Chem. Phys. Lett.* **1982**, *89*, 399-404

Bertz: *J. Am. Chem. Soc.* **1981**, *103*, 3599-3601

```python
descriptor_df['Balaban J'] = [GraphDescriptors.BalabanJ(mol) for mol in mol_objects]
descriptor_df['Bertz CT'] = [GraphDescriptors.BertzCT(mol) for mol in mol_objects]

print(descriptor_df.shape)
descriptor_df.head()
```

##### 3D Descriptors¶
https://rdkit.org/docs/source/rdkit.Chem.Descriptors3D.html

We need to generate 3D coordinates for our molecules first!

For now let's work with a subset of RDKit-embdedded molecules. Normally we would want to generate 3D descriptors from structures optimized at some level of theory (using MM, semi-empirical, or DFT methods). We are also glossing over conformational analysis ...

```python
%%time 

# generate 3D structs for RDKit mols (this may take 5-10 minutes)
mol_3d = []
smi_3d = data_formatted['SMILES']
for i, mol in enumerate(tqdm(mol_objects)): 
    # add H atoms
    molH = Chem.AddHs(mol)
    # create the 3D coordinates
    # returns -1 if the embed process fails, so we will store the variable embed to check this
    embed = AllChem.EmbedMolecule(molH,useExpTorsionAnglePrefs=True,useBasicKnowledge=True,randomSeed=42)
    # you could add an optimization with MMFF or UFF here
    
    if embed != -1:
        mol_3d.append(molH)
    
    else:
        mol_3d.append(np.nan)
```

```python
# write to file if needed

'''
writer = Chem.SDWriter('mol_3D.sdf')

for n, mol in enumerate(mol_3d):
    if not pd.isnull(mol):
        mol.SetProp('ID', descriptor_df['SMILES'].tolist()[n]) # associated data
        writer.write(mol)
writer.close()
'''

# read from file if needed
'''
    mol_3d = Chem.SDMolSupplier('mol_3D.sdf')
    smi_3d = [mol.GetProp('ID') for mol in mol_3d]
'''
```

```python
# create a new dataframe for the 3D descriptors
descriptor_3d = pd.DataFrame()
descriptor_3d['SMILES'] = smi_3d
descriptor_3d['molobj'] = mol_3d

# alternatively you can try to create and add the molobjects directly to a df if you have a smiles column:
# from rdkit.Chem import PandasTools
# PandasTools.AddMoleculeColumnToFrame(descriptor_3d,'SMILES','Molecule'

original_shape = descriptor_3d.shape
```

```python
# lets remove those molecules for which we were unable to embed a valid structure
descriptor_3d = descriptor_3d[descriptor_3d['molobj'].notna()]
new_shape = descriptor_3d.shape
print(' Failed to embed {} molecules'.format(original_shape[0] - new_shape[0]))
print(descriptor_3d.shape)
```

```python
# Add some 3D descriptors
descriptor_3d['Asphericity'] = [Descriptors3D.Asphericity(mol) for mol in descriptor_3d['molobj']]
descriptor_3d['Radius of Gyration'] = [Descriptors3D.RadiusOfGyration(mol) for mol in descriptor_3d['molobj']]
descriptor_3d['PMI'] = [Descriptors3D.PMI1(mol) for mol in descriptor_3d['molobj']]# First (smallest) principal moment of inertia
descriptor_3d = descriptor_3d.drop(['molobj'], axis=1)
descriptor_3d.head()
```

```python
# lets combine the 2D and 3D descriptors, bearing in mind that we weren't able to obtain 3D descriptors for all molecules
# the merged df will match the length of the smaller of the two 

# while we are at it, lets also include the y-value for our problem, the mean complexity score
y_vals = data[['SMILES','meanComplexity']]
merged = descriptor_df.merge(descriptor_3d,on='SMILES').merge(y_vals,on='SMILES')

merged.to_csv('merged.csv')
```

#### 6. Visualizing the data
We have obtained 8 molecular descriptors (one of the columns is the SMILES, one is the y value) for 1667 structures.

It's hard to visualize how the molecules differe in this 8 dimensional space, and so we will now apply dimensionality reduction techniques:

(a) Principal Component Analysis (PCA): a linear method
(b) t-distributed stochastic neighbor embedding (t-SNE): a non-linear method

Many others are available such as UMAP. 

These are "unsupervised learning" techniques. They compress the data by finding a smaller, different set of variables that capture what matters most in the original features, while minimizing the loss of information.

```python
# the y values
y = merged['meanComplexity']

# the molecular descriptors
numeric_data = merged.select_dtypes(include=['number'])
X = numeric_data.drop(['meanComplexity'], axis=1)
```

```python
# Feature scaling through standardization (or Z-score normalization) is an important preprocessing step 
# for many machine learning algorithms. Standardization involves rescaling the features such that they 
# have the properties of a standard normal distribution with a mean of zero and a standard deviation of one. 

scaler = preprocessing.StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

##### 6a. Perform PCA and plot the data distribution in terms of the 2 first principal components

```python
pca = PCA(8)  # construct linear combinations of descriptors to create new features, 
              # which are ordered according to how much of the sample variance is accounted for

projected = pca.fit_transform(X_scaled)
print(projected.shape)
```

```python
# Color each data point by the y-value to see how this is distributed in the descriptor space.
plt.rcParams['axes.grid'] = False

plt.scatter(projected[:, 0], projected[:, 1],
            c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.colorbar();
```

##### Are 2 principal components enough - i.e., do they account for the sample variance?

```python
pca = PCA().fit(X_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_.round(2)))
```

```python
# OK so try a 3D plot ...
pca = PCA(3)  # project from 8 to 2 dimensions
projected = pca.fit_transform(X_scaled)
print(projected.shape)

ax = plt.figure(figsize=(16,10)).gca(projection='3d')
ax.scatter(
    xs=projected[:, 0],
    ys=projected[:, 1],
    zs=projected[:, 2],
    c=y, 
    cmap=plt.cm.get_cmap('Spectral', 10)
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
plt.show()
```

##### 6b. Create a t-SNE plot

```python
perplexity = np.arange(10, 210, 10)
divergence = []

for i in tqdm(perplexity):
    model = TSNE(n_components=2, init="pca", perplexity=i)
    reduced = model.fit_transform(X)
    divergence.append(model.kl_divergence_)
    
plt.plot(perplexity, divergence)
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=175, n_iter=300)
tsne_results = tsne.fit_transform(X)
```

```python
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],
            c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.xlabel('TSNE component 1')
plt.ylabel('TSNE component 2')
plt.colorbar();
```

```python
# K-means clustering is a popular unsupervised machine learning algorithm 
# used for partitioning a dataset into a pre-defined number of clusters

# if this gives an error it may be necessary to update the following:
# pip install --upgrade threadpoolctl

for n_clusters in np.arange(2, 10):
    clusterer = KMeans(n_clusters=n_clusters)
    preds = clusterer.fit_predict(projected)
    centers = clusterer.cluster_centers_

    score = silhouette_score(projected, preds)
    print("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(projected)
y_kmeans = kmeans.predict(projected)
```

```python
plt.figure(figsize=(8,8))
# Let's visualize the results by plotting the data
plt.scatter(projected[:, 0], projected[:, 1], c=y_kmeans, s=30, cmap='viridis', alpha=0.5)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=400, alpha=0.3)
```
